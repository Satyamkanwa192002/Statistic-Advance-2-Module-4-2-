```python
#(i) Mean and Median Functions

def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:
        return sorted_numbers[mid]
```

```python
#(ii) Mode Function

from collections import Counter

def calculate_mode(numbers):
    frequency = Counter(numbers)
    mode = [key for key, value in frequency.items() if value == max(frequency.values())]
    return mode if len(mode) > 1 else mode[0]
```

```python
#(iii) Weighted Mean Function

def calculate_weighted_mean(values, weights):
    total_weighted_values = sum(value * weight for value, weight in zip(values, weights))
    total_weights = sum(weights)
    return total_weighted_values / total_weights
```

```python
#(iv) Geometric Mean Function

import math

def calculate_geometric_mean(numbers):
    product = 1
    for number in numbers:
        product *= number
    return product ** (1 / len(numbers))
```

```python
#(v) Harmonic Mean Function

def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / number for number in numbers)
```

```python
#(vi) Midrange Function

def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2
```

```python
#(vii) Trimmed Mean Function

def calculate_trimmed_mean(numbers, trim_percent):
    sorted_numbers = sorted(numbers)
    trim_count = int(len(numbers) * trim_percent / 100)
    trimmed_numbers = sorted_numbers[trim_count:-trim_count]
    return sum(trimmed_numbers) / len(trimmed_numbers)
```

```python
#(ii) Range Function

def calculate_range(numbers):
    return max(numbers) - min(numbers)
```

```python
#(iii) Variance and Standard Deviation Functions

def calculate_variance(numbers):
    mean = calculate_mean(numbers)
    return sum((x - mean) ** 2 for x in numbers) / len(numbers)

def calculate_standard_deviation(numbers):
    return calculate_variance(numbers) ** 0.5
```

```python
#(iv) Interquartile Range (IQR) Function

def calculate_iqr(numbers):
    sorted_numbers = sorted(numbers)
    q1 = calculate_median(sorted_numbers[:len(sorted_numbers) // 2])
    q3 = calculate_median(sorted_numbers[(len(sorted_numbers) + 1) // 2:])
    return q3 - q1
```

```python
#(v) Coefficient of Variation Function

def calculate_coefficient_of_variation(numbers):
    mean = calculate_mean(numbers)
    std_dev = calculate_standard_deviation(numbers)
    return (std_dev / mean) * 100
```

```python
#(vi) Mean Absolute Deviation (MAD) Function

def calculate_mad(numbers):
    mean = calculate_mean(numbers)
    return sum(abs(x - mean) for x in numbers) / len(numbers)
```

```python
#(vii) Quartile Deviation Function

def calculate_quartile_deviation(numbers):
    iqr = calculate_iqr(numbers)
    return iqr / 2
```

```python
#(viii) Range-Based Coefficient of Dispersion Function

def calculate_range_coefficient_of_dispersion(numbers):
    range_value = calculate_range(numbers)
    return range_value / (max(numbers) + min(numbers))
```

Q3. Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance.

1. Python Class for a Discrete Random Variable
This class will have methods to calculate the expected value (mean) and variance of a discrete random variable based on given probabilities for each outcome.

```python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if sum(probabilities) != 1:
            raise ValueError("Probabilities must sum to 1.")
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum((value - mean) ** 2 * prob for value, prob in zip(self.values, self.probabilities))

# Example usage
values = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6
die = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", die.expected_value())
print("Variance:", die.variance())
```

2. Simulating the Rolling of a Fair Six-Sided Die
We’ll simulate rolls of a six-sided die and calculate the expected value and variance of the outcomes.

```python
import random

def simulate_die_rolls(num_rolls):
    outcomes = [random.randint(1, 6) for _ in range(num_rolls)]
    mean = sum(outcomes) / num_rolls
    variance = sum((x - mean) ** 2 for x in outcomes) / num_rolls
    return mean, variance

# Simulate 10,000 rolls
mean, variance = simulate_die_rolls(10000)
print("Simulated Mean:", mean)
print("Simulated Variance:", variance)
```

Q5. Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance

3. Generating Random Samples from a Given Probability Distribution
This function will generate random samples from a specified distribution (e.g., binomial or Poisson) and calculate the mean and variance of the samples.

```python
import numpy as np

def generate_samples(distribution, params, sample_size):
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, sample_size)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, sample_size)
    else:
        raise ValueError("Unsupported distribution")

    mean = np.mean(samples)
    variance = np.var(samples)
    return samples, mean, variance

# Example for binomial distribution
samples, mean, variance = generate_samples('binomial', (10, 0.5), 1000)
print("Binomial Distribution - Mean:", mean, "Variance:", variance)

# Example for Poisson distribution
samples, mean, variance = generate_samples('poisson', (3,), 1000)
print("Poisson Distribution - Mean:", mean, "Variance:", variance)
```

Q8. Write a Python function to calculate the probability density function (PDF) of a continuous random
variable for a given normal distribution.


```python
import math

def normal_pdf(x, mean, std_dev):
    """Calculate the PDF of a normal distribution at a given point x."""
    variance = std_dev ** 2
    exponent = math.exp(-((x - mean) ** 2) / (2 * variance))
    return (1 / (math.sqrt(2 * math.pi * variance))) * exponent

# Example usage
mean = 0
std_dev = 1
x = 1
pdf_value = normal_pdf(x, mean, std_dev)
print(f"The PDF value of x={x} for N({mean}, {std_dev}^2) is {pdf_value}")
```

Q9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

```python
import math

def exponential_cdf(x, rate_parameter):
    """Calculate the CDF of an exponential distribution at a given point x."""
    if x < 0:
        return 0  # CDF is 0 for x < 0 in an exponential distribution
    return 1 - math.exp(-rate_parameter * x)

# Example usage
rate_parameter = 0.5  # Example rate parameter (lambda)
x = 2
cdf_value = exponential_cdf(x, rate_parameter)
print(f"The CDF value of x={x} for an exponential distribution with λ={rate_parameter} is {cdf_value}")
```

Q10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

```python
import math

def poisson_pmf(k, rate_parameter):
    """Calculate the PMF of a Poisson distribution at a given point k."""
    if k < 0 or not isinstance(k, int):
        return 0  # PMF is 0 for negative or non-integer k values
    return (rate_parameter ** k * math.exp(-rate_parameter)) / math.factorial(k)

# Example usage
rate_parameter = 3  # Average rate (lambda)
k = 2  # Number of occurrences
pmf_value = poisson_pmf(k, rate_parameter)
print(f"The PMF value for k={k} in a Poisson distribution with λ={rate_parameter} is {pmf_value}")
```

Q15. A company is testing two different website layouts to see which one leads to higher click-through rates.
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.

```python
import numpy as np
from scipy import stats

def ab_test(layout_a, layout_b):
    # Calculate means and standard deviations
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    std_a = np.std(layout_a, ddof=1)
    std_b = np.std(layout_b, ddof=1)
    n_a = len(layout_a)
    n_b = len(layout_b)

    # Calculate t-statistic
    t_statistic = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

    # Calculate degrees of freedom
    df = ((std_a**2 / n_a + std_b**2 / n_b)**2) / \
         (((std_a**2 / n_a)**2 / (n_a - 1)) + ((std_b**2 / n_b)**2 / (n_b - 1)))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

    return t_statistic, df, p_value

# Given data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Run the A/B test
t_statistic, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)
t_statistic, df, p_value
```

Q16. A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in
reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the tstatistic and p-value for the treatment effect.

```python
import numpy as np
from scipy import stats

def drug_effectiveness_test(existing_drug, new_drug):
    # Calculate means and standard deviations
    mean_existing = np.mean(existing_drug)
    mean_new = np.mean(new_drug)
    std_existing = np.std(existing_drug, ddof=1)
    std_new = np.std(new_drug, ddof=1)
    n_existing = len(existing_drug)
    n_new = len(new_drug)

    # Calculate t-statistic
    t_statistic = (mean_existing - mean_new) / np.sqrt((std_existing**2 / n_existing) + (std_new**2 / n_new))

    # Calculate degrees of freedom
    df = ((std_existing**2 / n_existing + std_new**2 / n_new)**2) / \
         (((std_existing**2 / n_existing)**2 / (n_existing - 1)) + ((std_new**2 / n_new)**2 / (n_new - 1)))

    # Calculate p-value (one-tailed test)
    p_value = 1 - stats.t.cdf(t_statistic, df)

    return t_statistic, df, p_value

# Given data
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Run the test
t_statistic, df, p_value = drug_effectiveness_test(existing_drug_levels, new_drug_levels)
t_statistic, df, p_value
```

Q17. A school district introduces an educational intervention program to improve math scores. Write a Python
function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
determine if the intervention had a significant impact.

```python
import numpy as np
from scipy import stats

def intervention_effectiveness_test(pre_scores, post_scores):
    # Calculate differences
    differences = np.array(post_scores) - np.array(pre_scores)
    
    # Mean and standard deviation of the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    n = len(differences)

    # t-statistic
    t_statistic = mean_diff / (std_diff / np.sqrt(n))

    # Degrees of freedom
    df = n - 1

    # p-value (one-tailed test)
    p_value = 1 - stats.t.cdf(t_statistic, df)

    return t_statistic, df, p_value

# Given data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Run the test
t_statistic, df, p_value = intervention_effectiveness_test(pre_intervention_scores, post_intervention_scores)
t_statistic, df, p_value
```

Q18. An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
significant difference between the average salaries of male and female employees.

```python
import numpy as np
from scipy import stats

def gender_salary_gap_test(male_salaries, female_salaries):
    # Calculate means and standard deviations
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)
    std_male = np.std(male_salaries, ddof=1)
    std_female = np.std(female_salaries, ddof=1)
    n_male = len(male_salaries)
    n_female = len(female_salaries)

    # Calculate t-statistic
    t_statistic = (mean_male - mean_female) / np.sqrt((std_male**2 / n_male) + (std_female**2 / n_female))

    # Calculate degrees of freedom
    df = ((std_male**2 / n_male + std_female**2 / n_female)**2) / \
         (((std_male**2 / n_male)**2 / (n_male - 1)) + ((std_female**2 / n_female)**2 / (n_female - 1)))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

    return t_statistic, df, p_value

# Generate synthetic salary data
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Run the test
t_statistic, df, p_value = gender_salary_gap_test(male_salaries, female_salaries)
t_statistic, df, p_value
```

```python
import numpy as np
from scipy import stats

def satisfaction_comparison_test(branch_a, branch_b):
    # Calculate means and standard deviations
    mean_a = np.mean(branch_a)
    mean_b = np.mean(branch_b)
    std_a = np.std(branch_a, ddof=1)
    std_b = np.std(branch_b, ddof=1)
    n_a = len(branch_a)
    n_b = len(branch_b)

    # Calculate t-statistic
    t_statistic = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

    # Calculate degrees of freedom
    df = ((std_a**2 / n_a + std_b**2 / n_b)**2) / \
         (((std_a**2 / n_a)**2 / (n_a - 1)) + ((std_b**2 / n_b)**2 / (n_b - 1)))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

    return t_statistic, df, p_value

# Given data
branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

# Run the test
t_statistic, df, p_value = satisfaction_comparison_test(branch_a_scores, branch_b_scores)
t_statistic, df, p_value
```

Q21. A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Output the results
chi2_stat, dof, p_value
```
